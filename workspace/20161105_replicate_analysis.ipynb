```python
import itertools

import protease_experimental_analysis
import protease_experimental_analysis.data as data
import protease_experimental_analysis.protease_sequencing_model as protease_sequencing_model
protease_sequencing_model = reload(protease_sequencing_model)
data = reload(data)
```

```python
def aligned_member_indicies(set_names):
    members = [data.model_input[n].values()[0]["name"] for n in set_names]
    return pandas.concat([
            pandas.Series(index=m, data=numpy.arange(len(m)), name = s)
             for s, m in zip(set_names, members)],
        axis=1, join="inner"
    )
```

```python
replicate_pairs = [
    ("rd2_chymo", "rd2_redo_chymo"),
    ("rd2_tryp", "rd2_redo_tryp"),
    ("rd3_chymo", "rd3_redo_chymo"),
    ("rd3_tryp", "rd3_redo_tryp"),
] 
```

```python
data_sets = list(itertools.chain(*replicate_pairs))
```

```python
models = {
    n : (
        protease_sequencing_model.FractionalSelectionModel(leak_prob_mass=False)
        .build_model(data.model_input[n]))
    for n in data_sets 
}
```

```python
%%time
params = { n : m.find_MAP() for n, m in models.items() }
```

```python
summary_frames = {}

for pair in replicate_pairs:
    sframes = []
    for n in pair:
        sf = data.counts[n].copy()
        sf["ec50"] = params[n]["sel_ec50"]
        sframes.append(sf)
    
    summary_frames[pair] = pandas.merge(*sframes, how="inner", on="name")
```

```python
for (x, y), f in summary_frames.items():
    jp = seaborn.jointplot(
        "ec50_x", "ec50_y", data=f,
        kind="hex", joint_kws = dict(mincnt=1)
    )
    jp.ax_joint.set_xlabel(x)
    jp.ax_joint.set_ylabel(y)
```
