```python
import protease_experimental_analysis.data as data
import protease_experimental_analysis.protease_sequencing_model as protease_sequencing_model
protease_sequencing_model = reload(protease_sequencing_model)
```

```python
model_sets = list(itertools.product(
        ("rd3_chymo", "rd3_redo_chymo"),
        (True, False)
    ))
```

```python
models = [
    protease_sequencing_model.FractionalSelectionModel(leak_prob_mass=leak)
    .build_model(data.model_input[dataset])
    for dataset, leak in model_sets
]
```

```python
fit_params = [m.find_MAP() for m in models]
```

```python
results = dict(zip(model_sets, fit_params))
```

```python
pickle.dump(results, open("161026_leak_analysis_params.pickle", "w"))
```

```python
sample_names = { m : data.model_input[m][0]["name"] for m in set(zip(*model_sets)[0]) }
```

```python
ec50_summary = pandas.concat(axis=1, join="inner", objs =[
    pandas.Series(name="%s_%s" % (m, l), data=v["sel_ec50"], index=sample_names[m])
    for (m, l), v in results.items()
])
```

```python
seaborn.jointplot(
    "rd3_redo_chymo_True", "rd3_redo_chymo_False", ec50_summary,
    kind="hex", joint_kws={"mincnt" : 1})

pylab.figure()

seaborn.jointplot(
    "rd3_chymo_True", "rd3_chymo_False", ec50_summary,
    kind="hex", joint_kws={"mincnt" : 1})
```

```python
seaborn.jointplot(
    "rd3_redo_chymo_False", "rd3_chymo_False", ec50_summary,
    kind="hex", joint_kws={"mincnt" : 1})

pylab.figure()

seaborn.jointplot(
    "rd3_redo_chymo_True", "rd3_chymo_True", ec50_summary,
    kind="hex", joint_kws={"mincnt" : 1})
```

```python
for n, m, p in zip(model_sets, models, fit_params):
    llh_summary = m.model_outlier_summary(p)
    sllh = [v["sel_log_likelihood_signed"] for v in llh_summary.values()]
    
    pylab.figure(figsize=(10, 3))
    pylab.title(n)
    pylab.boxplot([s[numpy.logical_and(s != 0, numpy.isfinite(s))] for s in sllh], showfliers=False)
    pylab.ylim((-150, 150))
```
